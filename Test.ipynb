```python
from DataRetreiver import Data_retreiver
```

```python
# read the API Key for Census Data API
# key can be requested at https://api.census.gov/data/key_signup.html
with open('API_KEY.txt', 'r') as file:
    api_key = file.read()
```

```python
# run the test case

xmin=-77.32
ymin=38.83
xmax=-77.24
ymax=38.88

test=Data_retreiver(xmin=xmin,ymin=ymin,xmax=xmax,ymax=ymax,var=['B01001_001E','B09002_001E'],
                    year=2020,dname='acs/acs5',key=api_key)
test
```

```python
# visualize the result using Folium
# the red box with dash line is the input bounding box(Area of Interest)
# the chropleth map is the total population
import folium
from branca.colormap import linear

# the default format for output column is pandas.object. For visualization, convert to int
test['B01001_001E']=test['B01001_001E'].astype(str).astype(int)

m = folium.Map(((ymin+ymax)/2, (xmin+xmax)/2), tiles="cartodb positron", zoom_start=12)

colormap = linear.YlGn_09.scale(
    test['B01001_001E'].min(), test['B01001_001E'].max()
)

folium.GeoJson(
    test,
    name="Total Population",
    style_function=lambda feature: {
        "fillColor": colormap(feature["properties"]["B01001_001E"]),
        "color": "black",
        "weight": 1,
        "fillOpacity": 0.9,
    },
).add_to(m)

kw = {
    "color": "red",
    "line_cap": "round",
    "fill": False,
    "weight": 3,
}

colormap.add_to(m)

folium.Rectangle(
    bounds=[[ymin, xmin], [ymax, xmax]],
    line_join="round",
    dash_array="5, 5",
    **kw,
).add_to(m)


m
```
